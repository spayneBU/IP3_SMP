### Problem 1
Write a function that:
- Takes the filename
- Converts the Date column to a date
- Add a new column called "Company" that contains the Company name

```python
import numpy as np
import pandas as pd
import csv
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
def convert_csv(file):
    df = pd.read_csv(file)
    df.insert(1,'Company',file[:-4])
    return df
```

### Problem 2
Execute the convert_csv( ) function four times, passing each file to it

```python
df1 = convert_csv('50Hertz.csv')
df2 = convert_csv('Amprion.csv')
df3 = convert_csv('TenneTTSO.csv')
df4 = convert_csv('TransnetBW.csv')
```

### Problem 3
Stack the dataframes on top of each to produce one large dataframe that has 98 columns, one for company, one for date, and one for each timepoint

```python
df = pd.concat([df1,df2,df3,df4],axis=0)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, yearfirst=False, format='%d/%m/%Y',errors='ignore')
df
```

```python
df.dtypes
```

### Problem 4
Create a new dataframe that is grouped by Company and the year

```python
df_grouped = df.groupby(['Company',df['Date'].dt.year]).mean()
df_grouped
```

### Problem 6
- Combine the two bar graphs into a single graph that shares the x-axis
- The plots should be in grayscale
- Save the file as 'combined_bar_chart.png'

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4)
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
df_grouped_midnight.sort_values().plot(kind='barh',title='TW/hours by Company/Year @ Midnight',ax=ax1,color='grey',sharex='col')
df_grouped_noon.sort_values().plot(kind='barh',title='TW/hours by Company/Year @ Noon',ax=ax2,color='dimgray')
fig.savefig('combined_bar_chart.png')
```

### Problem 7
Write a function that accepts the ungrouped dataframe, the company name, date, and a plotting color that generates an area graph for that particular company/day

Use this function signature: *daily_view(dataframe, company, date, color)*

```python
def daily_view(dataframe, company, date, color):
    df_area = dataframe[(dataframe['Company']==company) & (dataframe['Date']==date)]
    df_area.drop(['Company','Date'], axis=1, inplace=True)
    df_area_transposed = df_area.T
    area_plot = df_area_transposed.plot.area(title='Power Production: {}/{}'.format(company,date), xlabel='Time of Day', ylabel='TW/h Generated', color=color, legend=None)
    return area_plot
```

```python
daily_view(df,'50Hertz','2019-08-24','salmon')
```
